```python
import pandas as pd
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
dir(digits)
```

```python
import matplotlib.pyplot as plt
plt.gray()
for i in range(3):
    plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis='columns'), df.target, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

#EXERCISE

Use famous iris flower dataset from sklearn.datasets to predict flower species using random forest classifier.

1. Measure prediction score using default n_estimators (10)
2. Now fine tune your model by changing number of trees in your classifer and tell me what best score you can get using how many trees

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
iris.data[:5]
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df["target"] = iris.target
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis='columns'), df.target, test_size=0.4)
```

```python
model = RandomForestClassifier(n_estimators=10)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
highest_score = 0
best_estimators = 0

for i in range(1,40):
    model = RandomForestClassifier(n_estimators=i)
    model.fit(X_train, y_train)
    print(i, model.score(X_test, y_test))
    if model.score(X_test, y_test) > highest_score:
        highest_score = model.score(X_test, y_test)
        best_estimators = i

print("Best amount of estimators :", best_estimators, highest_score)
```

```python
y_predicted = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predicted)
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
