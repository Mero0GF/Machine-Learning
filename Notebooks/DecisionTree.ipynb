```python
import pandas as pd
df = pd.read_csv("salaries.csv")
df.head()
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_new = inputs.drop(['company','job','degree'],axis='columns')
inputs_new
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_new, target)
```

```python
model.score(inputs_new,target)
```

Prediction:\
google - sales executive - masters\
google - computer programmer - masters\
facebook - sales executive - degree

```python
model.predict([[2,2,1],[2,1,1],[1,2,0]])
```

Based on titanic.csv, find if a person survived the incident based on their Pclass, Sex, Age and Fare

```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
len(df["Age"])
```

```python
df["Age"].isna().sum()
```

```python
df["Age"] = df["Age"].dropna()
```

```python
inputs = df.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
target = df['Survived']
```

```python
inputs.head()
```

```python
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
```

```python
inputs["Sex_n"] = le_Sex.fit_transform(inputs["Sex"])
```

```python
inputs.head()
```

```python
inputs_new = inputs.drop(["Sex"],axis="columns")
```

```python
inputs_new.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(inputs_new,target,test_size=0.5,random_state=1130)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
